```python
#10828
import sys

N=int(sys.stdin.readline())

stack=[]
pos=0

def push(x, pos):
    stack.append(x)
    pos+=1
    
    return pos
def pop(pos):
    if pos > 0:
        print(stack.pop(pos-1))
        pos-=1
    else:
        print(-1)
    
    return pos

def size(pos):
    print(pos)

def empty(pos):
    if pos==0:
        print(1)
    else:
        print(0)
        
    return pos

def top(pos):
    if pos==0:
        print(-1)
    else:
        print(stack[pos-1])
        
    return pos

for i in range(N):
    
    string=input().split(" ")
    if len(string)>1:
        A, B= map(str,string)
    else:
        A=string[0]

    if A=="push":
        pos=push(int(B), pos)
    elif A=="pop":
        pos=pop(pos)
    elif A=="size":
        size(pos)
    elif A=="empty":
        pos=empty(pos)
    elif A=="top":
        pos=top(pos)
    
```

```python
# 1874 다시풀기
import sys
stack=[]

n=int(sys.stdin.readline().rstrip())

rest=[i for i in range(1,n+1)]
stack=[]
result=[]

for _ in range(n):
    n_=int(sys.stdin.readline().rstrip())

    while(1):
        
        
        if len(rest)==0:
            break
        else:
            pop=rest.pop(len(rest)-1)
            if n_==pop:
                break
            else: stack.append(pop)
            print("+")
        print(rest, stack, result)

    result.append(pop)
    print("-")
    print(rest, stack, result)
```
